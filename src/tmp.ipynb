```python
df_sum = df.groupby(['source']).sum().reset_index()

# Ogni 4 righe c'è una nuova colonna
# Prendo i 4 nodi, calcolo il totale della colonna, ogni nodo avrà come valore il 
# punto medio del suo intervallo e parto da 1 sottraendo

tmp_df = df_sum.copy()
tmp_df = tmp_df.loc[tmp_df['source'] < 4]

tot = sum(tmp_df.head(4)['value'])

tmp_df["pos_tmp"] = tmp_df['value'] / tot
tmp_df["pos"] = 1 - (tmp_df["pos_tmp"].cumsum() - tmp_df["pos_tmp"] / 2)
tmp_df
```
